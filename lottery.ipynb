```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.taiwanlottery.com.tw/index_new.aspx'
```

```python
resp=requests.get(url)
resp
```

```python
soup=BeautifulSoup(resp.text,'lxml')
soup
```

### 字典化

```python
lottery_group={
    '威力彩':2,
    '大樂透':4,
    '金彩539':6    
}
lottery_group
```

```python
main=soup.find('div',id=f"contents_logo_0{lottery_group['大樂透']}").find_parent('div')
main
```

```python
date=main.find('span').text.strip().replace('\xa0',' ')
date
```

```python
balls=[ball.text.strip() for ball in main.find_all('div',class_="ball_tx")[6:]]
balls
```

```python
if main.find('div',class_='ball_red'):
    balls=balls+[main.find('div',class_='ball_red').text.strip()]
    
balls
```

```python
def get_ball(name):
    main=soup.find('div',id=f"contents_logo_0{lottery_group[name]}").find_parent('div')
    date=main.find('span').text.strip().replace('\xa0',' ')
    balls=','.join([ball.text.strip() for ball in main.find_all('div',class_="ball_tx")[6:]])
    if main.find('div',class_='ball_red'):
        balls=balls+' 特別號:'+main.find('div',class_='ball_red').text.strip()
        
    return date,balls
```

```python
get_ball('威力彩')
```

## 迴圈格式輸出

```python
datas={}

for name in lottery_group:
    datas[name]=get_ball(name)
    
datas
```
